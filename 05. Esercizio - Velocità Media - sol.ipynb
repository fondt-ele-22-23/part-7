**La velocità media su una sezione $[x_0, x_1]$ è data da:**

$$
\bar{v} = \frac{1}{x_1 - x_0} \int_{x_0}^{x_1} v(x) \, dx
$$


Nel modulo `sol.condotta`, si definisca per tale calcolo la funzione:

```python
def media(f, x0, x1)
```

* I parametro `x0` e `x1` rappresentano i valori di $x_0, x_1$
* Il parametro `f` è una funzione con parametro `x` che indica la velocità alla posizione `x`
* Si utilizzi per l'integrazione la funzione `scipy.integrate.quad`
* La funzione deve restituire solo la velocità media (non il bound sull'errore)
* Si calcoli la velocità media nella sezione $[0, 3.7]$
* Il definisca il codice di collaudo nella cella seguente

````python
vm = condotta.media(condotta.v, x0=0, x1=3.7)
print(f"La velocità media nella condotta è: {vm:.3f}")
````
